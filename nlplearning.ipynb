IN THIS WE ARE CLEANING THE TEXT (cleaning the input)
step 3: TEXT PREPROCESSING 3->Advanced Vectorization (Word2vec, avgword2vec) vectors are used to create meaningfull insigths about the data or text which further can be used in machine learning models
